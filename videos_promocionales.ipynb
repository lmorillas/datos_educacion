```python
import folium
from folium import plugins

import pandas as pd
import branca
```

```python
# json con datos de centros y v√≠deos
centros = pd.read_json('videos_promocionales.json')

CENTRO = (41.6561, -0.8773)
```

```python
mapa = folium.Map(location=CENTRO, zoom_start=12)


fg=folium.FeatureGroup(name="V√≠deos promocionales")

# fondo blanco
style_function = lambda x: {'fillColor': '#ffffff'}

# silueta de Arag√≥n
folium.GeoJson('ara.json', name='aragon', style_function=style_function).add_to(mapa)


html_video = '''<h2>{}</h2>
<iframe width="427" height="240" src="https://www.youtube.com/embed/{}" frameborder="0" gesture="media" allow="encrypted-media" rel="0" allowfullscreen></iframe>'''
html_video = '''<iframe width="427" height="240" src="https://www.youtube.com/embed/{}" frameborder="0" gesture="media" allow="encrypted-media" rel="0" allowfullscreen></iframe>'''

for lat,lon,name,video in zip(centros['latitude'], centros['longitude'], centros['centro'], centros['video']):
    iframe = branca.element.IFrame(html = html_video.format(video), width=440, height=260)
    fg.add_child(folium.Marker(location=[lat,lon],tooltip=name, icon=folium.Icon(icon='facetime-video'),
        popup=folium.Popup(iframe, max_width=440)))
    
mapa.add_child(fg)

mapa
```
